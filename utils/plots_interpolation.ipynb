```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Loading CSV
df = pd.read_csv("/Users/sujandevkota/Desktop/Thesis/thesis2025/FINAL_RESULTS/Final_Combined_Results_CLEAN.csv")


interp_df = df[df["Method"] == "Interpolation"].copy()


interp_df["ResValue"] = interp_df["ResValue"].astype(int)
interp_df = interp_df.sort_values("ResValue")


plt.figure(figsize=(6, 4))
sns.lineplot(
    data=interp_df,
    x="ResValue",
    y="mAP@0.5:0.95",
    hue="Test Set",  
    marker="o",
    palette="muted"
)

plt.title("Interpolation: mAP@0.5:0.95 on Synthetic vs. Native Test Sets")
plt.xlabel("Input Resolution")
plt.ylabel("mAP@0.5:0.95")
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()


plt.savefig("interpolation_map_50_95.png", dpi=300)

plt.show()
```
